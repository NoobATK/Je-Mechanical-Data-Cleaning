```python
get_ductboard_purchases = materials_wo_co
cleaned_materials_to_concat = materials_wo_co[~(materials_wo_co['Item'].str.contains('ductboard|ductboard r-4', case=False) & ~materials_wo_co['Item'].str.contains('ductboard sheet|ductboard start', case=False))]
get_ductboard_purchases[['Length','Width']] = materials_wo_co['Item'].str.extract(r'^(\d{1,2})x(\d{1,2})')
get_ductboard_purchases = get_ductboard_purchases[get_ductboard_purchases['Item'].str.contains('ductboard', case=False)]

ductb_purchases_values = pd.DataFrame()
# group by measurement and r value
if not get_ductboard_purchases.empty:
    get_ductboard_purchases.loc[:,'R-Value'] = get_ductboard_purchases['Item'].apply(lambda x: 'r-6' if re.search(r'r[-\s]?6', x, flags=re.IGNORECASE) else ('r-8' if re.search(r'r[-\s]?8', x, flags=re.IGNORECASE) else None))
    ductb_purchases_values = get_ductboard_purchases.groupby(['Length','Width', 'R-Value'], as_index=False).agg({'QTY Received': 'sum', 'Price Billed': 'sum'})
    ductb_purchases_values['Total Sheets'] = 0
```

```python
def ductb_calculator(df):
    for index, rows in df.iterrows():
        total_in_box = (int(rows['Length'])*2)+(int(rows['Width'])*2)
        if rows['R-Value'] == 'r-6':
            sheet_cost = 42.96
            total_in_box += 12
        elif rows['R-Value'] == 'r-8':
            sheet_cost = 62.32
            total_in_box += 16
        if total_in_box < 120:
            num_box1_cut = 1
        in_residual = 120-total_in_box
        if in_residual >= total_in_box:
            num_box2_cut = 1
        elif in_residual >=(total_in_box/2):
            num_box2_cut = 0.5
        else:
            num_box2_cut = 0 
        in_residual_2 = (120-total_in_box) - (total_in_box*num_box2_cut)
        total_box_sheet = num_box1_cut + num_box2_cut
        qty = rows['QTY Received']
        total_sheets_required = qty/total_box_sheet
        df.at[index, 'Total Sheets'] = total_sheets_required
        residual_sqr_in = in_residual_2*48
        min_cap_sqr_inc = 14*12
        num_of_caps = residual_sqr_in/min_cap_sqr_inc
        total_cost_x_size = total_sheets_required*sheet_cost
        if qty > 0:
            box_unit_cost = total_cost_x_size/qty
            labor_cost_x_box = 0
            box_unit_total_cost = box_unit_cost+labor_cost_x_box
            total_cost = box_unit_total_cost*qty
            df.at[index, 'Total Cost'] = total_cost
        else: 
            df.at[index, 'Total Cost'] = 0

ductb_calculator(ductb_purchases_values)
```

Funcion de separacion de Lineset

```python
def lineset_separator(row,rows,copper_1,copper_2,armaflex,value,armavalue):
    items = [f'{value}" copper roll 100ft', '3/8" copper roll 100ft', f'{value}" x {armavalue}" armaflex 100ft']
    qty = row['QTY'] / 2
    item_price = row['Price']
    vendor = row['Vendor']
    prices = [item_price*copper_1,item_price*copper_2,item_price*armaflex]
    for item in items:
        rows.append({'Item': item, 'QTY': qty,'Vendor':vendor})
    for i, price in enumerate(prices):
        rows[-len(items) + i]['Price'] = round(price,2)
```

```python
takeoff_grouping_items = []
takeoff_grouping_unit_conv = []
takeoff_reference_dict = {}

# Convierte la base de datos en un diccionario para poder iterar las keys que tengan (.)
for document in collection_takeoff_reference.find():
    takeoff_reference_dict[document['Item']] = document['TakeOff Items']

# Compara los items de la tabla de takeoff para agrupar, unificar y subir a la base de datos.
for index, row in to_materials_w_lineset.iterrows():
    item = row['Item']
    # Desempaca el diccionario para acceder a los takeoff items y luego empatar todos con el "Item principal" del diccionario
    for item_de_empate, takeoff_item_desempacado in takeoff_reference_dict.items():
        for takeoff_de_empate, unit_converter_dict in takeoff_item_desempacado.items():
            for unit, unit_value in unit_converter_dict.items():
                if item == takeoff_de_empate:
                    takeoff_grouping_items.append(item_de_empate)
                    takeoff_grouping_unit_conv.append(unit_value)
                    break
            else:
                continue
            break
        else:
            continue
        break
    else:
        takeoff_grouping_items.append(f'{item} (Not in TakeOff)')
        takeoff_grouping_unit_conv.append(1)

to_materials_w_lineset['Grouping_Item'] = takeoff_grouping_items
to_materials_w_lineset['unit_converter'] = takeoff_grouping_unit_conv
to_materials_w_lineset['Grouping_QTY'] = to_materials_w_lineset['QTY']* to_materials_w_lineset['unit_converter'] 
unified_takeoff_table = to_materials_w_lineset.groupby('Grouping_Item').agg({'Grouping_QTY':'sum','Total Price':'sum','Price':'mean'}).reset_index()
```

```python
def purchased_out_of_total_takeoff(df):
    for index, rows in df.iterrows():
        if rows['ESTIMATE IN TAKEOFF'] > rows['Total QTY Received']:
            df.at[index, 'PURCHASED OUT OF TOTAL TAKEOFF'] = rows['Total QTY Received']
        else:
            df.at[index, 'PURCHASED OUT OF TOTAL TAKEOFF'] = rows['ESTIMATE IN TAKEOFF']

def prct_qty_not_estimated(df,row1,row2,newrow):
    for index,rows in df.iterrows():
        denominator = rows[row1]
        numerator = rows[row2]
        if denominator == 0:
            df.at[index,newrow] = 1
        else:
            df.at[index,newrow] = ((numerator/denominator)-1)*-1
```

```python
def get_col_widths(dataframe):
    # First, we find the maximum length of the index column and the headers of each column
    idx_max = max([len(str(s)) for s in dataframe.index.values] + [len(str(dataframe.index.name))])
    header_max = max([len(str(s)) for s in dataframe.columns.values])
    
    # Then, we iterate through each column and determine the maximum length of the data in that column.
    # We add 2 to account for padding spaces on either side of each column
    # Finally, we return a list of the maximum lengths of each column.
    return [max(idx_max, header_max)+2] + [max([len(str(s)) for s in dataframe[col].values] + [len(col)]) for col in dataframe.columns]

excel_output = project_name

writer = pd.ExcelWriter(f'C:\\Users\\jecoo\\Desktop\\Alejandro\\Programming\\mongodb\\finals\\{excel_output}_final.xlsx', engine='xlsxwriter')
# Convert the dataframe to an XlsxWriter Excel object. Turn off the default
# header and index and skip one row to allow us to insert a user defined
# header.

dataframes = [purchases_materials, to_materials_w_lineset,materials_no_extras ,purchase_takeoff_table,materials_returned_grouped,direct_labor,purchased_extras,extras_grouped]
datasheets = ['Purchases', 'Takeoff_Materials','Purchase_comparison','Materials_Breakdown','Returned_Material','Direct_Cost_Labor','Extras','Extras_Grouped']

format_dict = {'Purchases': {'Price Billed': '$#,##0.00'},
               'Takeoff_Materials': {'Price': '$#,##0.00','QTY':'$#,##0.00','Total Price':'$#,##0.00'},
               'Materials_Breakdown': {'Price Billed': '$#,##0.00'}}

for i, df in enumerate(dataframes):
    # Get the sheet name and create an empty dictionary for the format strings.
    sheet_name = datasheets[i]
    formats = {}
    if sheet_name in format_dict:
        formats = format_dict[sheet_name]

    # Write the dataframe to the sheet.
    df.to_excel(writer, sheet_name=sheet_name, startrow=1, header=False, index=False)

    # Get the workbook and worksheet objects.
    workbook = writer.book
    worksheet = writer.sheets[sheet_name]

    # Get the dimensions of the dataframe.
    (max_row, max_col) = df.shape

    # Create a list of column headers, to use in add_table().
    column_settings = []
    for header in df.columns:
        col_format = formats.get(header, None)
        if col_format:
            column_settings.append({'header': header, 'format': workbook.add_format({'num_format': col_format})})
        else:
            column_settings.append({'header': header})

    # Add the table.
    worksheet.add_table(0, 0, max_row, max_col - 1, {'columns': column_settings, 'name': datasheets[i]})

    # Make the columns wider for clarity.
    for a, width in enumerate(get_col_widths(df)):
        worksheet.set_column(a-1, a-1, width)

ps_dropped_ret_dfl.to_excel(writer, sheet_name='Subcontractors', index=False, startrow=0, startcol=0)
retencion.to_excel(writer, sheet_name='Subcontractors', index=False, startrow=len(ps_dropped_ret_dfl) + 2, startcol=0)
subcontractor_merged.to_excel(writer, sheet_name='Subcontractors', index=False, startrow=0, startcol=ps_dropped_ret_dfl.shape[1] + 2)
subcontractors_table.to_excel(writer, sheet_name='Subcontractors', index=False, startrow=len(subcontractor_merged) + 2, startcol=ps_dropped_ret_dfl.shape[1] + 2)


eq_mat_excel.to_excel(writer, sheet_name='Equipment', index=False, startrow=0, startcol=0)
takeoff_equipment.to_excel(writer, sheet_name='Equipment', index=False, startrow=0, startcol=eq_mat_excel.shape[1] + 2)
equipment_table_df.to_excel(writer, sheet_name='Equipment', index=False, startrow=len(takeoff_equipment) + 2, startcol=eq_mat_excel.shape[1] + 2)

exhaustfan_materials.to_excel(writer, sheet_name='Exhaust Fans', index=False, startrow=0, startcol=0)
takeoff_exhausfan.to_excel(writer, sheet_name='Exhaust Fans', index=False, startrow=0, startcol=exhaustfan_materials.shape[1] + 2)
exhaust_fan_df.to_excel(writer, sheet_name='Exhaust Fans', index=False, startrow=len(takeoff_exhausfan) + 2, startcol=exhaustfan_materials.shape[1] + 2)

role_dict_row = 0
role_dict_col = 0
for index,df in role_dict.items():
    df.to_excel(writer, sheet_name='Direct Field Labor', index=False, startrow=role_dict_row, startcol=0)
    role_dict_row += len(df)+2
    role_dict_col = df.shape[1]

dfl_detailed.to_excel(writer, sheet_name='Direct Field Labor', index=False, startrow=0, startcol=role_dict_col + 2)
merged_time_grouped.to_excel(writer, sheet_name='Direct Field Labor', index=False, startrow=len(dfl_detailed) + 2, startcol=role_dict_col + 2)



# Close the Pandas Excel writer and output the Excel file.
writer.close()
```
